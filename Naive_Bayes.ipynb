## Gaussian Naive Bayes

Gaussian Naive Bayes (GNB) is a classification technique used in machine learning based on a probabilistic approach and Gaussian distribution. Gaussian Naive Bayes assumes that each parameter, also called features or predictors, has an independet capacity of predicting the output variable.

Naive Bayes classifiers are supervised machine learning algorithms used to solve classification problems. The term 'Naive' comes from the fact that Naive Bayes model assume all features in a model are independet, meaning they do not affect one another.

# Task: 

Create a Classifier model for the iris.csv utilizing a Naive Bayes algorithm

## Load the dataset

```python
import pandas as pd
```

```python
df=pd.read_csv('iris.csv')
df
```

### Illustrative iris flower image

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

img=mpimg.imread('iris_types.png')
plt.figure(figsize=(10,24))
plt.axis('off')
plt.imshow(img)
```

### Selecting data into independent and dependent variables

```python
X = df.iloc[:,:4].values
y = df['species'].values
```

```python
X,y
```

### Split 80/20

```python
from sklearn.model_selection import train_test_split
#splitting into train set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.20, random_state =82)

```

## Standarization

Standard scaling transforms numerical features to have a mean of 0 and a standard deviation of 1, making it suitable for algorithms that assume normally distributed data. 

Advantages:

- Handles features with different units well
- Reduces the impact of outliers

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)

X_test = sc. transform(X_test)
```

```python
X_train
```

## Built the Naive Bayes Classification Model

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
```

### Training the model

```python
model.fit(X_train, y_train)
```

## Predictions

```python
predictions = model.predict(X_test)
```

```python
predictions
```

## Accuracy

```python
from sklearn import metrics
```

```python
accuracy = metrics.accuracy_score(y_test, predictions)
accuracy
```
